````python
from openai import OpenAI
import os
from dotenv import load_dotenv
prompt="Translate the following English sentence into french: 'The weather is nice today'"

load_dotenv()
api=os.getenv('openai_key')
print(api)
client = OpenAI(api_key=api)

response = client.chat.completions.create(model= "gpt-4o-mini",
   messages=[{"role":"user","content":prompt}],temperature=0)

print(response.choices[0].message.content)
````

Multi step prompting

````python
text= "Biola is a key part of the group of fiends.He has travelled out of the country last year"

prompt= f"""You will be provided with a text delimited by triple backticks.
step1: Identify the key entities in the sentence.
step2: Classify each entity as a person ,event or object. ```{text}```"""

client = OpenAI(api_key=api)
response= client.chat.completions.create(model="gpt-4o-mini",messages=[{"role":"user","content":prompt}],temperature=0)

print(response.choices[0].message.content)
````
