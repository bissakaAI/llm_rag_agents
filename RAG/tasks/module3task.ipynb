```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Sample sentences
sentences = [
    "The dog is playing in the park",
    "A puppy is running outside",
    "The cat is sleeping on the couch",
    "Python is a programming language",
    "Machine learning models need data",
    "I love coding in Python"
]

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Generate embeddings
embeddings = model.encode(sentences)

# Function to compute similarity against all others
def similarity_against_all(query_index):
    query_embedding = embeddings[query_index].reshape(1, -1)
    similarities = cosine_similarity(query_embedding, embeddings)[0]
    return similarities

# Similarity for sentence 1 and sentence 4
sim_sentence_1 = similarity_against_all(0)
sim_sentence_4 = similarity_against_all(3)
```
