```python
import os 
from dotenv import load_dotenv

load_dotenv()
openai_api_key = os.getenv("openai_key")

if not openai_api_key:
    print("Warning: OpenAikey not found.Set it in env file")

else:
    print("API key loaded successfully")
```

```python
documents = [ "Python is highl-level programming languages known for readeability and simplicity.","Python is widely used in web development, data science, artificial intelligence, scientific computing,and automation. Popular frameworks include Django and Flask for",
             "One of Python's key strengths is its extensive standard library, which provides tools for many common programming tasks."]
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from openai import OpenAI


model= SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings= model.encode(documents)
print(doc_embeddings)

#2. Query and retrieve 
query="what is one key strength of python?"
query_embedding= model.encode([query])[0]
```

```python
# 3. compute similarity 
similarities=np.dot(doc_embeddings,query_embedding)
top_idx= np.argmax(similarities)
retrieved_doc= documents[top_idx]
```

```python
retrieved_doc
```

```python
#4. Generate response 
client=OpenAI(api_key=openai_api_key)
prompt= f""" Context:{retrieved_doc}

Answer based on the context"""

response= client.chat.completions.create(model="gpt-3.5-turbo",messages=[{"role":"user","content":prompt}],temperature=0)
```
